```python
from collections import defaultdict, Counter

ids_per_interest = defaultdict(list)
for user_id, interest in interests:
    ids_per_interest[interest].append(user_id)

interests_per_id = defaultdict(list)
for user_id, interest in interests:
    interests_per_id[user_id].append(interest)

def most_common_interests_with(user):
    return Counter(
        interested_user_id 
        for interest in interests_per_id[user['id']]
        for interested_user_id in ids_per_interest[interest]
        if interested_user_id != user['id']
    )

most_common_interests_with({ 'id': 0 })
```

```python
class CountClicker:

    def __init__(self, default_count=0):
        self.count = default_count

    def click(self, count=1):
        self.count += count

    def read(self):
        return self.count
    
    def reset(self):
        self.count = 0


clicker = CountClicker()
clicker.click()
clicker.click()
clicker.click()
print(clicker.read())
clicker.reset()
print(clicker.read())
clicker.click()
clicker.click()
print(clicker.read())
```

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256] 
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1] 
total_error  = [x + y 
for x, y in zip(variance, bias_squared)] 
xs = [i for i, _ in enumerate(variance)]
fig, ax = plt.subplots()
ax.plot(xs, variance, color='g', linestyle=':', label='variance')
ax.plot(xs, bias_squared, color='r', linestyle='--', label='bias^2')
ax.plot(xs, total_error, color='b', linestyle='-', label='total error')
ax.set_xlabel("model complexity")
ax.set_ylabel("MSE")
ax.legend()
plt.show()
```

```python
from collections import namedtuple
import datetime
StockPrice = namedtuple('StockPrice', ['symbol', 'date', 'closing_price'])
price = StockPrice('MSFT', datetime.date(2002, 8, 10), 106.3)
assert price.symbol == 'MSFT'
assert price.closing_price == 106.3
print(price.symbol, price.date, price.closing_price)
```

```python
from typing import NamedTuple
class StockkPrice(NamedTuple):
    symbol: str
    date: datetime.date
    closing_price: float

    def is_high_tech(self) -> bool:
        return self.symbol in ['MSFT', 'GOOG', 'FB', 'AMZN', 'AAPL']
    
price = StockkPrice('TNT', datetime.date(2018, 12, 14), 454.1)
assert price.symbol == 'TNT'
assert price.closing_price == 454.1
assert not price.is_high_tech()
```

```python
from dataclasses import dataclass 
@dataclass 
class StockPrice2: 
    symbol: str 
    date: datetime.date 
    closing_price: float 
    def is_high_tech(self) -> bool: 
        """It's a class, so we can add methods too""" 
        return self.symbol in ['MSFT', 'GOOG', 'FB', 'AMZN', 'AAPL'] 

price2 = StockPrice2('MSFT', datetime.date(2018, 12, 14), 106.03) 
assert price2.symbol == 'MSFT' 
assert price2.closing_price == 106.03 
assert price2.is_high_tech()
print(price2)
```
